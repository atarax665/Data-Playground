```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno
```

```python
# Source - https://www.mygov.in/corona-data/covid19-statewise-status/
data = pd.read_csv('D:/dataset/covid_india.csv')
```

```python
#Renaming colums with big names
data.rename(columns = {'Name of State / UT':'State', 'Active Cases':'Active', 
                       'Cured/Discharged/Migrated':'Cured', 'Total Confirmed cases':'Total'},inplace=True)
```

```python
data.head(5)
```

```python
# Visualizing Active cases
plt.figure(figsize=(20,10))
plt.xlabel('Name of State / UT', size=14, color='b')
plt.ylabel('Active Cases', size=14,color='r')
plt.plot(data['State'],data['Active'])
plt.xticks(rotation = 'vertical', color='indigo')
plt.yticks(range(0,360000,10000))
plt.title('Active Cases',size=24)
plt.grid (True)
```

```python
# Visualizing total number of deaths
plt.figure(figsize=(20,10))
plt.xlabel('Name of State / UT', size=14,color='b')
plt.ylabel('Deaths', size=14,color='r')
plt.plot(data['State'],data['Deaths'])
plt.xticks(rotation = 'vertical', color='indigo')
plt.yticks(np.arange(0,55000,2000))
plt.title('Deaths',size=24)
plt.grid (True)
```

```python
#Visualizing total confirmed cases
plt.figure(figsize=(18,10))
plt.bar(data["State"],data["Total"],edgecolor="black")
plt.title("Total Confirmed Cases",fontsize=15)
plt.xlabel("State / Union terirorry",fontsize=14)
plt.ylabel("Confirmed",fontsize=14)
plt.xticks(rotation=90,size=14)
plt.yticks(size=14)
plt.show()
```

```python
# Recovery rate will be percentage of cured patient out of total infected people
recovery_rate = data['Cured']/data['Total']*100
```

```python
plt.figure(figsize=(20,10))
plt.xlabel('State / Union terirorry', size=14,color='b')
plt.ylabel('Recovery Rate(%)', size=14,color='r')
plt.plot(data['State'],recovery_rate)
plt.xticks(rotation='vertical' , color='indigo')
plt.yticks(np.arange(85,101,1))
plt.title('Recovery Rate', size = 20)
plt.grid(True)
```

```python
# State wise visualization of data w.r.t active cases
data_sorted_act = data.sort_values(by = 'Active', ascending = False).reset_index(drop = True)
data_sorted_act.style.background_gradient(cmap = 'Reds')
```
